## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder #for label encoding
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

## Load data

```python

df = pd.read_excel('/worldcities.xlsx')
```

```python
df
```

 ## Add column

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این مرحله با توجه به ستون جمعیت یک ستون جدید تحت عنوان Population Level   اضافه می‌کنیم به طوری که به 4 دسته زیر  بر اساس وضعیت هر داده اختصاص داده می شودند.
</br>
    <div dir="ltr">
Low: 0-25%
    </br>
Mid: 25-50%
    </br>
High: 50-70%
    </br>
Over: 70-100%
    </div>
</font>
</p>
</div>

```python

Q25 = df['population'].quantile(0.25)
Q50 = df['population'].quantile(0.5)
Q70 = df['population'].quantile(0.70)

def popLvl_determination(pop):
  if pop < Q25:
    return 'Low'
  elif pop < Q50:
    return 'Mid'
  elif pop < Q70:
    return 'High'
  else:
    return 'Over'

for i in df.index:
  df.at[i, 'population level'] = popLvl_determination(df.at[i, 'population'])
```

```python
df
```

## Preproccess

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این بخش ستون های خالی داده ها راحذف کرده و در نهایت تمامی ستون ها را به داده های عددی تبدیل می‌کنیم به جز ستون اضافه شده
</font>
</p>
</div>

```python

df.dropna(inplace=True)
```

```python
df
```

```python
le = LabelEncoder()
df['ville'] = le.fit_transform(df['ville'])
df
```

```python
df['ville_ascii'] = le.fit_transform(df['ville_ascii'])
df['pays'] = le.fit_transform(df['pays'])
df['iso2'] = le.fit_transform(df['iso2'])
df['iso3'] = le.fit_transform(df['iso3'])
df['admin_nom'] = le.fit_transform(df['admin_nom'])
df['capital'] = le.fit_transform(df['capital'])
df
```

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 حال باید داده های به دست آمده از قسمت قبلی را نرمال سازی کنیم.
</font>
</p>
</div>

```python

for col in df.columns:
  if col != 'population level':
    min_col = df[col].min()
    max_col = df[col].max()
    df[col] = (df[col] - min_col) / (max_col - min_col)
```

```python
df.describe()
```

## Split

```python

train_size = int(0.8 * len(df))

train_set = df[:train_size]
test_set = df[train_size:]
test_set
```

## Regression

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای این قسمت باید ستون Population  را به عنوان برچسب درنظر گرفته و از بقیه ستون هابه عنوان فیچر استفاده کنیم
</font
</p>
</div>

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
از رگرسیون خطی
،درجه 1 و درجه 2 استفاده کرده و در هر مورد دقت و خطای میانگین مربعات را بدست آورده و نتایج را باهم مقایسه می‌کنیم.</font
</p>
    
</div>

```python

X_train = train_set.drop(['population', 'population level'], axis=1) # we didn't include population level because its not numeric
y_train = train_set['population']
X_test = test_set.drop(['population', 'population level'], axis=1)
y_test = test_set['population']
X_test
```

```python
# First-degree linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)

mse_first_degree = mean_squared_error(y_test, y_pred)
acc_first_degree = lin_reg.score(X_test, y_test)

print("First Degree Linear Regression:")
print("MSE:", mse_first_degree)
print("acc:", acc_first_degree)

```

```python
# Second-degree polynomial regression
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_train_poly, y_train)
y_pred_poly = lin_reg_poly.predict(X_test_poly)

mse_second_degree = mean_squared_error(y_test, y_pred_poly)
acc_second_degree = lin_reg_poly.score(X_test_poly, y_test)

print("\nSecond Degree Polynomial Regression:")
print("MSE:", mse_second_degree)
print("acc:", acc_second_degree)
```

```python
# Comparing the results
if mse_first_degree < mse_second_degree:
    print("\nFirst degree model is better because it has a lower MSE.")
else:
    print("\nSecond degree model is better because it has a lower MSE.")

if acc_first_degree > acc_second_degree:
    print("First degree model is better because it has a higher accuracy.")
else:
    print("Second degree model is better because it has a higher accuracy.")
```

## Classification

### Decision Tree (Entropy)

```python


X_train = train_set.drop(['population', 'population level'], axis=1)
y_train = train_set['population level']
X_test = test_set.drop(['population', 'population level'], axis=1)
y_test = test_set['population level']

y_test
```

```python
# Creating and training the decision tree classifier
dtc = DecisionTreeClassifier(criterion='entropy', max_depth=6) # for setting the max_depth we have to be carefull to avoid underfitting and overfitting.
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of the Decision Tree model: {:.2f}%".format(accuracy * 100))
```

### Random Forest (Entropy)

```python


rf_clf = RandomForestClassifier(criterion='entropy', max_depth=6)
rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of the Random Forest model: {:.2f}%".format(accuracy * 100))
```

```python
# as we can see from the result above, Random forrest works better than desicion tree and it makes sense
```

### KNN K=2

```python


knn2_clf = KNeighborsClassifier(n_neighbors=2)
knn2_clf.fit(X_train, y_train)

y_pred = knn2_clf.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the KNN (k=2) model: {accuracy * 100:.2f}%")
```

### KNN K=3

```python

knn3_clf = KNeighborsClassifier(n_neighbors=3)
knn3_clf.fit(X_train, y_train)

y_pred = knn3_clf.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the KNN (k=3) model: {accuracy * 100:.2f}%")
```

### KNN K=5

```python

knn5_clf = KNeighborsClassifier(n_neighbors=5)
knn5_clf.fit(X_train, y_train)

y_pred = knn5_clf.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the KNN (k=5) model: {accuracy * 100:.2f}%")
```

```python
# in the KNNs, as we can see, the K=3 had the best results but all of them are weaker classifiers compared to dtc and rf.
```

### SVM (Linear)

```python


svm_clf = SVC(kernel='linear', random_state=42)
svm_clf.fit(X_train, y_train)

# Predicting
y_pred = svm_clf.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the SVM (linear) model: {accuracy * 100:.2f}%")
```

### SVM (Non-linear)

```python

svm_clf = SVC(kernel='rbf', random_state=42)
svm_clf.fit(X_train, y_train)

# Predicting
y_pred = svm_clf.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the SVM (RBF kernel) model: {accuracy * 100:.2f}%")
```

```python
# as u can see due to accuracy of both linear and non linear svm, a non linear is a better choice with attention to our dataset.
```
